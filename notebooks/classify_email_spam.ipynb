### *Classify Email Spam with Logistic Regression*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fake data (number of times the words "Present" and "Pay" appear in the emails)
X = np.array([
    [2, 3], [3, 2], [0, 1], [0, 0], [1, 3], [5, 5], [0, 2]
])
# y = 1: Spammed email, y = 0: not spammed
y = np.array([1, 1, 0, 0, 1, 1, 0]).reshape(-1, 1)

# init parameters
W = np.zeros((2, 1))
b = 0

# Set Hyperparameter
learning_rate = 1e-3
epochs = 10000
```

### *Sigmoid Function*
$$\sigma = \frac{1}{1 + \exp{}^{-z}}$$

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

### *Binary Cross Entropy*
$$Loss = -\frac{1}{n} \sum [y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y})]$$

```python
def compute_loss(X, y, W, b):
    n = len(y)
    z = X@W + b
    y_pred = sigmoid(z)
    return -1 * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred)) / n
```

```python
print("--Training--")
for i in range(epochs):
    n = len(y)
    z = X@W + b
    y_pred = sigmoid(z)

    dW = ((X.T @ (y_pred - y))) / n 
    db = np.sum((y_pred - y)) / n 

    W = W - learning_rate*dW
    b = b - learning_rate*db

    current_loss = compute_loss(X, y, W, b)
    print(f"LOSS: {current_loss:.4f}")
    
```

```python
print("--Predict--")
T = np.zeros((1, 2))
labels = ["present", "pay"]
for i in range(T.shape[1]):
    T[0, i] = float(input(f"Enter times of {labels[i]}"))
prediction = sigmoid(T@W + b)
print(prediction)
if prediction[0, 0] > 0.5:
    print("Spammed!!")
else:
    print("Not Spammed!!")
```
